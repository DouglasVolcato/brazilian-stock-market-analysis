```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
import pandas as pd
import time
```

```python
path = 'chromedriver-win64.exe'

options = Options()
options.headless = True
options.add_argument("--headless=new")

service = Service(executable_path=path)
driver = webdriver.Chrome(service=service, options=options)
```

```python
stocks = pd.read_csv('b3_stocks.csv')['nome']
```

```python
def get_page_value(item):
    try:
        container = driver.find_element("xpath", f'//td[contains(.,"{item}")]')
        item = container.find_element("xpath", './following-sibling::td')
        return item.text
    except:
        return '0'
```

```python
data = []
for index,stock in enumerate(stocks):
    website = f'https://www.fundamentus.com.br/detalhes.php?papel={stock}&h=1'
    driver.get(website)
    time.sleep(1)

    data.append({
        'nome': stock,
        'Tipo': get_page_value('Tipo'),
        'Empresa': get_page_value('Empresa'),
        'Setor': get_page_value('Setor'),
        'Subsetor': get_page_value('Subsetor'),
        'Cotação': get_page_value('Cotação'),
        'Valor de mercado': get_page_value('Valor de mercado'),
        'Valor da firma': get_page_value('Valor da firma'),
        'P/L': get_page_value('P/L'),
        'P/VP': get_page_value('P/VP'),
        'P/EBIT': get_page_value('P/EBIT'),
        'PSR': get_page_value('PSR'),
        'P/Ativos': get_page_value('P/Ativos'),
        'P/Cap. Giro': get_page_value('P/Cap. Giro'),
        'P/Ativ Circ Liq': get_page_value('P/Ativ Circ Liq'),
        'Div. Yield': get_page_value('Div. Yield'),
        'EV / EBITDA': get_page_value('EV / EBITDA'),
        'EV / EBIT': get_page_value('EV / EBIT'),
        'Cres. Rec (5a)': get_page_value('Cres. Rec (5a)'),
        'LPA': get_page_value('LPA'),
        'VPA': get_page_value('VPA'),
        'Marg. Bruta': get_page_value('Marg. Bruta'),
        'Marg. EBIT': get_page_value('Marg. EBIT'),
        'Marg. Líquida': get_page_value('Marg. Líquida'),
        'EBIT / Ativo': get_page_value('EBIT / Ativo'),
        'ROIC': get_page_value('ROIC'),
        'ROE': get_page_value('ROE'),
        'Liquidez Corr': get_page_value('Liquidez Corr'),
        'Div Br/ Patrim': get_page_value('Div Br/ Patrim'),
        'Giro Ativos': get_page_value('Giro Ativos'),
        'Ativo': get_page_value('Ativo'),
        'Patrim. Líq': get_page_value('Patrim. Líq')
    })

    print(f'{((index + 1) * 100) / len(stocks)}% - stock: {stock}')

```

```python
stocks_data = pd.DataFrame(data)
```
